ENCODING TECHNIQUES LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Gender = le.fit_transform(df.Gender)
df.head()
```

X AND Y SPLIT

```python
# independent varibles-X

X=df.drop(columns=['Spending Score (1-100)'],axis=1)
X.head()
```

```python
# y target-dependent variable


y=df['Spending Score (1-100)']
y
```

SCALING

```python
from sklearn.preprocessing import scale
X_scaled=pd.DataFrame(scale(X),columns=X.columns)
X_scaled.head()
```

CLUSTERING ALGORITHM

```python
from sklearn import cluster
error =[]
for i in range(1,11):
    kmeans=cluster.KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X_scaled)
    error.append(kmeans.inertia_)
```

```python
error
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),error)
plt.title('Elbow method')
plt.xlabel('no of clus')
plt.ylabel('error')
plt.show()
```

```python
km_model=cluster.KMeans(n_clusters=3,init='k-means++',random_state=0)
km_model.fit(X_scaled)
```

```python
ykmeans =km_model.predict(X_scaled)
ykmeans
```

```python
km_model.predict([[5.1,3.5,1.4,0.2]])
```

```python
X_scaled['kclus'] = pd.Series(ykmeans) # Adding the cluster data with the primary dataset
X_scaled.head()
```

Split the data into dependent and independent
variables

```python
x=X_scaled.iloc[:,:-1]
x                           # Split the data into dependent and independent variables.
```

```python
Y=X_scaled.kclus
```

Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size=0.2,random_state = 1)
```

Build the model 

```python
# model building 
from sklearn.ensemble import RandomForestClassifier
model =RandomForestClassifier(n_estimators=15,max_depth=3)
```

```python
#Fit the model
model.fit(x_train,y_train)
```

```python
test_pred = model.predict(x_test)
train_pred = model.predict(x_train)
```

Evaluate the model

```python
# evaluating the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

Train the model

```python
print('Training accuracy: ',accuracy_score(y_train,train_pred))
```

Test the model

```python
print('Testing accuracy: ' , accuracy_score(y_test,test_pred))
```

Evaluation metrics

```python
pd.crosstab(y_test,test_pred)
```
